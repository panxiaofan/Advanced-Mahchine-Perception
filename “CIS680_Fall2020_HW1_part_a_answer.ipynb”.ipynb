# 1)  Plot Loss and Gradient

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
min=-5;max=5;num=21;
w=np.linspace(min,max,num)
b=np.linspace(min,max,num)
w_grid,b_grid=np.meshgrid(w,b)
wt=torch.from_numpy(w_grid);bt=torch.from_numpy(b_grid)
x=torch.Tensor([1])
y_hat=torch.sigmoid(wt*x+bt)

########################### Q1.1 ###############################################
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(w_grid, b_grid, y_hat.numpy())
ax.set_xlabel('weight'); ax.set_ylabel('bias'); ax.set_zlabel('Sigmoid function');ax.set_title("Sigmoid_Activation")
plt.draw()
plt.savefig('Sigmoid_Activation.png')
```

```
########################### Q1.2 ###############################################
y=torch.Tensor([0.5])
l2=torch.square(y_hat-y)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(w_grid, b_grid, l2.numpy())
ax.set_xlabel('weight'); ax.set_ylabel('bias'); ax.set_zlabel('L2 Loss');ax.set_title("L2 Loss")
plt.draw()
plt.savefig('L2 Loss.png')
```

```
########################### Q1.3 ###############################################
dl2dy_hat=2*(y_hat-y)
dy_hatdz=y_hat*(torch.Tensor([1])-y_hat)
dzdw=x
dl2dw=dl2dy_hat * dy_hatdz * dzdw
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(w_grid, b_grid, dl2dw.numpy())
ax.set_xlabel('weight'); ax.set_ylabel('bias'); ax.set_zlabel('L2 Loss');ax.set_title("gradien of L2 Loss w.r.t. weights")
plt.draw()
plt.savefig('gradien of L2 Loss w.r.t. weights.png')
```

```
########################### Q1.4 ###############################################
y=torch.Tensor([0.5])
lce=-y*torch.log(y_hat)+(y-torch.Tensor([1]))*torch.log(torch.Tensor([1])-y_hat)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(w_grid, b_grid, lce.numpy())
ax.set_xlabel('weight'); ax.set_ylabel('bias'); ax.set_zlabel('Cross Entropy Loss');ax.set_title("Cross Entropy Loss")
plt.draw()
plt.savefig('Cross Entropy Loss.png')
```

```
########################### Q1.5 ###############################################
dlcedy_hat=-y/y_hat + (torch.Tensor([1])-y)/(torch.Tensor([1])-y_hat)
dy_hatdz=y_hat*(torch.Tensor([1])-y_hat)
dzdw=x
dlcedw=dlcedy_hat * dy_hatdz * dzdw
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(w_grid, b_grid, dlcedw.numpy())
ax.set_xlabel('weight'); ax.set_ylabel('bias'); ax.set_zlabel('L2 Loss');ax.set_title("gradien of Cross Entropy Loss w.r.t. weights")
plt.draw()
plt.savefig('gradien of Cross Entropy Loss w.r.t. weights.png')
```
